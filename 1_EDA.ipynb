```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

print("Train shape:", train.shape)
print("Test shape:", test.shape)
print("\nTrain columns:", train.columns.tolist())

print("\nMa'lumotlar haqida umumiy ma'lumot:")
print(train.info())

print("\nStatistik ma'lumotlar:")
print(train.describe())

plt.figure(figsize=(10, 6))
sns.histplot(train['yield'], kde=True)
plt.title('Hosildorlik taqsimoti')
plt.xlabel('Hosildorlik')
plt.ylabel('Soni')
plt.show()

important_features = ['clonesize', 'honeybee', 'bumbles', 'fruitmass']

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
for i, feature in enumerate(important_features):
    row = i // 2
    col = i % 2
    sns.scatterplot(data=train, x=feature, y='yield', ax=axes[row, col])
    axes[row, col].set_title(f'Yield vs {feature}')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 8))
sns.heatmap(train.corr(), annot=True, cmap='coolwarm', center=0)
plt.title('Korrelyatsiya matritsasi')
plt.show()

pollinators = ['honeybee', 'bumbles', 'andrena', 'osmia']
plt.figure(figsize=(10, 6))
train[pollinators].boxplot()
plt.title('Changlatuvchilar taqsimoti')
plt.ylabel('Miqdori')
plt.xticks(rotation=45)
plt.show()

temp_cols = ['MinOfUpperTRange', 'MaxOfUpperTRange', 'MinOfLowerTRange', 'MaxOfLowerTRange']
plt.figure(figsize=(10, 6))
train[temp_cols].boxplot()
plt.title('Harorat taqsimoti')
plt.ylabel('Harorat')
plt.xticks(rotation=45)
plt.show()
```
